```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def formatAcct(acct):
    acctName = acct.split("/raw/")[1]
    return acctName.split(".csv")[0]
```

```python
def plot(tmp):
    # Plots should all be straight lines because the ratio of raw:prepped is a constant
    fig, ax = plt.subplots(figsize=(12, 3))
    raw = tmp.iloc[:,0]
    norm = tmp.iloc[:,1]
    plt.plot(raw, norm)
    plt.show()
    plt.close()
```

##### Make sure the prepped values match the raw data

```python
rawLoc = '/home/tbrownex/data/retirement/funds/raw/*'
normLoc = '/home/tbrownex/data/retirement/funds/normalized/'
```

```python
for fund in glob.glob(rawLoc):
    fundName = formatAcct(fund)
    print(fundName)
    raw = pd.read_csv(fund)
    raw.set_index('Date', inplace=True)
    fName = normLoc+fundName+'.csv'
    norm = pd.read_csv(fName)
    norm.set_index('date', inplace=True)
    tmp = raw.join(norm, how='inner')
    tmp = tmp[['Close', fundName]]
    plot(tmp)
```
