##### Get all the Symbols that MarketStack has, for Indexes

```python
import requests
import json

from utils import formatURL
```

```python
typ = 'exchanges/INDX/tickers'
url = formatURL(typ)
```

```python
resp = requests.get(url).json()

l = []
for idx in resp['data']['indexes']:
    l.append((idx['name'], idx['symbol']))

for idx in sorted(l, key=lambda x: x[0]):
    print("{:<40}{}".format(idx[0], idx[1]))
```
