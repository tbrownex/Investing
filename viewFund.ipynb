```python
import pandas as pd
from plotColumns import plotColumns

from config import getConfig
```

```python
cfg = getConfig()
df = pd.read_csv(cfg['dataLoc']+'fundValues.csv')
```

```python
df.set_index('date', inplace=True)
df.sort_index(inplace=True)
```

```python
symbol = 'QYLD'
```

```python
fund = df[df['fund']==symbol]['fundVal']
benchMark = df[df['fund']=='SP500TR.INDX']['price']
fund = fund[fund.index.isin(benchMark.index)]
benchMark = benchMark[benchMark.index.isin(fund.index)]

plotColumns(fund, benchMark, symbol, 'SPX')
```
